```python
import pandas as pd

def read_files(filenames):
    dataframes = []
    for filename in filenames:
        tmp = pd.read_csv(filename, sep=';', usecols=[0, 1, 2, 3], header=None, names=['id', 'demand', 'supply', 'timestamp'])
#         tmp['demand'] = tmp['demand'].astype(str)
        dataframes.append(tmp)
    return dataframes

def split_string_to_array(field):
    field = field[0:-1]
    tmp = field.split("'")[1::2]
    return [float(i) for i in tmp]
```

```python
class Currency_pair:
    def __init__(self, supply, demand, timestamp):
        self._supply = supply
        self._demand = demand
        self._timestamp = timestamp
        
    @property
    def supply(self):
        return self._supply
    
    @property
    def demand(self):
        return self._demand
    
    @property
    def timestamp(self):
        return self._timestamp
```
