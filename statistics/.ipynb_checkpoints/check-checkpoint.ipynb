```python
import numpy as np
import json
import sys 
from os.path import isfile, join, isdir
from os import listdir
from typing import Dict, Tuple, Sequence
```

```python
path = './output_data/'
fee = 0.001
```

```python
def fetch_directories(path: str) -> [str]:
    '''
    Fetches all directories in the given path 
    '''
    directories = []
    try:
        directories = [join(path, f) for f in listdir(path) if isdir(join(path, f))] 
    except:
        print("Trouble opening a directory")
        return directories
    return directories

def fetch_files(path: str) -> [str]:    
    '''
    Fetches all files in the given path 
    '''
    files = []
    try:
        files = [f for f in listdir(path) if isfile(join(path, f))] 
    except:
        print("Trouble opening a directory")
        return files
    return files

            
def flatten(l: [[any]]) -> [any]:
    '''
    Takes a 2-D array and flattens it into a 1-D array
    '''
    return [item for sublist in l for item in sublist]
```

```python

def supply_calculation(opportunity: str, name: str):
    indexes = opportunity['supply_gain_index']
    calculated_score = 1 / opportunity['pairs'][0]['supply'][indexes[0]][0] * opportunity['pairs'][1]['demand'][indexes[1]][0] * opportunity['pairs'][2]['demand'][indexes[2]][0]
    calculated_score *= (1-fee)**3  # poplatky
    AAABBB = opportunity['pairs'][0]['supply'][indexes[0]]
    AAACCC = opportunity['pairs'][1]['demand'][indexes[1]]
    CCCBBB = opportunity['pairs'][2]['demand'][indexes[2]]
    breadth = (
                    opportunity['pairs'][0]['supply'][indexes[0]][1],
                    opportunity['pairs'][1]['demand'][indexes[1]][1],
                    opportunity['pairs'][2]['demand'][indexes[2]][1] 
                        * opportunity['pairs'][2]['demand'][indexes[2]][0]
                        / opportunity['pairs'][0]['supply'][indexes[0]][0]
                   )
    sfee = 1 - fee
    ret = ''
    if np.argmin(breadth) == 0:
        if False:
            print('SHISITITITIS1')
        a = AAABBB[1] * AAACCC[0] * sfee**2
        if a > CCCBBB[1]:
            print('a1')
            print('min:', min(breadth))
            print('new:', CCCBBB[1] / (a) * min(breadth)) 
            ret = name
    elif np.argmin(breadth) == 1:
        a = AAACCC[1] * AAACCC[0] * sfee
        if a > CCCBBB[1]:
            print('b1')
            print('min:', min(breadth))
            print('new:', CCCBBB[1] / (a) * min(breadth)) 
            ret = name
        a = AAACCC[1] * AAACCC[0] * CCCBBB[0] / AAABBB[0] * sfee**2
        if a > AAABBB[1]:
            print('c1')
            print('min:', min(breadth))
            print('new:', AAABBB[1] / (a) * min(breadth)) 
            ret = name
    else:
        a = CCCBBB[1] * CCCBBB[0] / AAABBB[0] * sfee
        if a > AAABBB[1]:
            print('d1')
            print('min:', min(breadth))
            print('new:', AAABBB[1] / (a) * min(breadth)) 
            ret = name
        a = CCCBBB[1] * CCCBBB[0] / AAABBB[0] * sfee**2
        if a > AAACCC[1]:
            print('e1')
            print('min:', min(breadth))
            print('new:', AAACCC[1] / (a) * min(breadth)) 
            ret = name
    return ret
    

def demand_calculation(opportunity: str, name: str):
    indexes = opportunity['demand_gain_index']
    calculated_score = opportunity['pairs'][0]['demand'][indexes[0]][0] / opportunity['pairs'][1]['supply'][indexes[1]][0] / opportunity['pairs'][2]['supply'][indexes[2]][0]
    AAABBB = opportunity['pairs'][0]['demand'][indexes[0]]
    AAACCC = opportunity['pairs'][1]['supply'][indexes[1]]
    CCCBBB = opportunity['pairs'][2]['supply'][indexes[2]]
#     print(AAABBB, AAACCC, CCCBBB, sep='\n')
    calculated_score *= (1-fee)**3  # poplatky
    breadth = (
                opportunity['pairs'][0]['demand'][indexes[0]][1],
                opportunity['pairs'][1]['supply'][indexes[1]][1],
                opportunity['pairs'][2]['supply'][indexes[2]][1] 
                    / opportunity['pairs'][1]['supply'][indexes[1]][0] * (1-fee)
                )
    sfee = 1 - fee
    ret = ''
    if np.argmin(breadth) == 0:
#         print(AAABBB[1]*AAABBB[0]/CCCBBB[0]*sfee, CCCBBB[1])
#         print(AAABBB[1]*AAABBB[0]/CCCBBB[0]/AAACCC[0]*sfee**2, AAACCC[1])
        a = AAABBB[1]*AAABBB[0]/CCCBBB[0]*sfee
        if  a > CCCBBB[1]:
            print('a')
            print('min:', min(breadth))
            print('new:', CCCBBB[1] / (a) * min(breadth))
            ret = name
        a = AAABBB[1]*AAABBB[0]/CCCBBB[0]/AAACCC[0]*sfee**2
        if  a > AAACCC[1]:
            print('b')
            print('min:', min(breadth))
            print('new:', AAACCC[1] / (a) * min(breadth))
            ret = name
    elif np.argmin(breadth) == 1:
        if False:
            print('FUUCUCUCK3')
        a = AAACCC[1]*AAABBB[0]/CCCBBB[0]*sfee**2
        if  a > CCCBBB[1]:
            print('c')
            print('min:', min(breadth))
            print('new:', CCCBBB[1] / (a) * min(breadth))
            ret = name
    else:
        a = CCCBBB[1]/AAACCC[0]
        if a > AAACCC[1]:
            print('d')
            print('min:', min(breadth))
            print('new:', AAACCC[1] / (a) * min(breadth))
            ret = name
        a = CCCBBB[1]/AAACCC[0]*sfee
        if CCCBBB[1]/AAACCC[0]*sfee > AAABBB[1]:
            print('e')
            print('min:', min(breadth))
            print('new:', AAABBB[1] / (a) * min(breadth))
            ret = name
    return ret
```

```python
wrong = []
def init():
#     if len(sys.argv) != 2:
#         print("Please enter an input directory path as a first argument with json file")
#         return 0;
    print('            Demand(*)        Supply(/)', 
    ' AAABBB       ->(*)            <-(/)',
    ' AAACCC       <-(/)            ->(*)',
    ' CCCBBB       <-(/)            ->(*)', sep='\n')
    data = ''
    for directory in fetch_directories(path):
        for file in fetch_files(directory):
#             if file != 'USDTEOSBNB-2020-03-12.json':
#                 continue
#             print(join(directory, file))
            with open(join(directory, file)) as f:
                data = json.loads(f.read())
            for opportunity in data['arbitrage_stats']:
                if float(opportunity['supply_gain']) > 0:
                    x = supply_calculation(opportunity, file)
                if float(opportunity['demand_gain']) > 0:
                    x = demand_calculation(opportunity, file)
                if x != '':
                    wrong.append(x)
    display(wrong)
init()
```
