```python
def get_narrowest(p1, p2, p3) -> float:
#     print(p1, p2, p3)
    narrowest = np.prod(p1)
    if narrowest > p2[1]:
        tmp = p2[1]
    if narrowest * p2[0] > p3[1]:
        a = p3[1] / p2[0] / p1[0]
        return a
    return narrowest / p1[0]

def get_max_gain(df):
    pairs = df['pairs']
    max_gain = 0
    for p1 in pairs[0]['supply']:
        for p2 in pairs[1]['demand']:
            for p3 in pairs[2]['demand']:
                score = p1[0]/p2[0]/p3[0]
#                 print("Score", score)
                if score > 1:
                    gain = get_narrowest(p1, p2, p3) * (score-1)
                    if max_gain < gain:
                        max_gain = gain
#                     if max_gain > 1:
#                         print(max_gain)
#                         print(p1, p2, p3)
    return max_gain
    
    return max_gain

gains = []
for d in df:
    gains.append(get_max_gain(d))
    
print("Total daily inefficiency ", sum(gains))
print("Max daily gain", max(gains))
```
